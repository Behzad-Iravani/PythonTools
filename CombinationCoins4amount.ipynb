This function computes the number of combinations of US coins for a give amount 



```python
def combination_coins(amount: int)->list:
    # Input:
    #       amount: integer
    # Output:
    #       a list of combination of coins 
    coin_list = [] # Initiazlize the output
    for amount_100 in range(0, amount+1,100):
        for amount_50 in range(0, amount+1,50):
            for amount_25 in range(0, amount+1,25):
                for amount_10 in range(0, amount+1,10):
                    for amount_5 in range(0, amount+1,5):
                        
                        total_value = amount_100 + amount_50 + amount_25 + amount_10 + amount_5
                        
                        if total_value <= amount:
                            coin_list.append([amount_100, amount_50 , amount_25, amount_10, amount_5, amount-total_value])
    return coin_list
```

```python
coin_list = combination_coins(100)
```

```python
print(len(coin_list))
```

Therefore there are 293 unique combinations of US coins for changing a 100$ bill

### Now we assess how the number of combination grows as function of total amount 

```python
len_comb = [len(combination_coins(amount)) for amount in range(100, 500, 25)]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
plt.plot(range(100, 500, 25), len_comb)
plt.xlabel('Total amount in US dollars')
plt.ylabel('Total number of combinations')
plt.grid('on')
```

This experiment indicated that there is exponetial growth in number of combinations as a function of total amount.
